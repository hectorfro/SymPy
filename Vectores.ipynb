```python
import sympy
from sympy import *
from sympy.vector import *
R = CoordSys3D('R')
```

```python
k,L,x,λ = symbols('k, L, x, λ')
```

```python
V=k*λ*log(( L+sqrt(L**2+4*R.x**2) )/(-L+sqrt(L**2+4*R.x**2)))
E=-gradient(V)
```

```python
E=simplify(E)
E
```

```python
curl(E)
```

```python
is_conservative(E)
```

```python
divergence(E).factor()
```
